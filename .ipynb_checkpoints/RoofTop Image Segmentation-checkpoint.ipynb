## Data Prepration

Since  we only have 25 training images, which is not enough to train a neural network, we will use ImageDataGenerator for data augmentation. Set seed=1 to make sure that the augmented images correspond to the right augmented labels. Data augmentation is the technique that increases the amount of data with some specific
methods which can be flipping, rotating, zooming, changing contrast and other characteristics of the image without damaging the content of the image. It is useful tool to increase the variety in the training images that the model will be trained on but using all transformation methods at the same time may reduce model performance so should be careful when it is used.

```python
# define the types of augmentations
from keras.preprocessing.image import ImageDataGenerator
aug_dict = dict(rotation_range=0.2,
                width_shift_range=0.05,
                height_shift_range=0.05,
                shear_range=0.05,
                zoom_range=0.05,
                horizontal_flip=True,
                fill_mode='nearest')

# define the right folders and 
batch_size=2 
train_path='./sample_data/train/'
#valid_path = './sample_data/valid/'
image_folder='images'
label_folder='labels'
image_color_mode='rgba'
label_color_mode='grayscale'
target_size=(256,256)
seed=1 # to make sure that the augmented images correspond to the right augmented labels. 


# define the data generators
image_datagen = ImageDataGenerator(**aug_dict)
label_datagen = ImageDataGenerator(**aug_dict)


image_generator = image_datagen.flow_from_directory(
    train_path,
    classes = [image_folder],
    class_mode = None,
    color_mode = image_color_mode,
    target_size = target_size,
    batch_size = batch_size,
    seed = seed)
label_generator = label_datagen.flow_from_directory(
    train_path,
    classes = [label_folder],
    class_mode = None,
    color_mode = label_color_mode,
    target_size = target_size,
    batch_size = batch_size,
    seed = seed)

train_generator = zip(image_generator, label_generator)
#valid_generator = zip(image_valid, mask_valid)


# With Generator function we can keep calling augmented images.
# This will be used when training the model.
# Transform the pixels in the label on the border of the roofs in the label
# to either 0 or 1. And if needed (just to be sure) resize the images.
def Generator(train_generator):
  for (img,mask) in train_generator:
      if(np.max(img) > 1): # just being sure to be scaled.
        img = img / 255.
        mask = mask /255.
      mask[mask > 0.5] = 1
      mask[mask <= 0.5] = 0
      yield (img,mask)
      
myGene = Generator(train_generator)
```

```python
# To check outputs of myGene
for example in range(1):
    example = next(myGene) # giving examples respectively
    for i in range(batch_size): # batch_size is 2
      img=example[0][i]
      label=example[1][i]
      fig, ax = plt.subplots(1, 2)
      ax[0].imshow(img.squeeze(), cmap=plt.get_cmap('bone')) # "bone" to see figure black and white
      ax[1].imshow(label.squeeze(), cmap=plt.get_cmap('bone'))
      fig.show()
```

# **MODEL ARCHITECTURE**

To see the an example of U-Net Structre Image : https://production-media.paperswithcode.com/methods/Screen_Shot_2020-07-07_at_9.08.00_PM_rpNArED.png

The main paper : https://arxiv.org/pdf/1505.04597v1.pdf



Classic computer vision approaches using deep learning solutions focused on classification.  It uses convolutional layers for feature learning, after that, densely-connected layers is added for assigning a class to the input image.

This allows you to distinguish between burger and pizza etc. Now, there is much more that, and it  is possible to draw boxes around objects present in an image using like Transformers.

However, it is possible to add even more precision to the images. While classic object detectors draw bounding boxes, image segmentation approaches perform at pixel level. Each pixel is inspected and assigned a class. This allows  to draw very sharp boxes separating the objects.

U-Net CNN is one of the main approches for that. And the inspired reference and basic model structure schem is given upper links.

The left part and foot of the U is called the contracting path and the right part is called the expansive path. Jointly, and with the help of skip connections, U-nets downsample an input image to learn about its evident features, to reconstruct the input (or a derived product, like a segmentation mask) via upsampling.



Now, Lets look at three main substructure of U-Net architecure.

**Contracting**:

In the work on U-Net, Ronneberger et al. (2015) architecture started with a regular convolutional neural network. Each convolution networks is contracting network what the paper call. This means that the convolutional layers (and pooling layers) scale down the feature maps (outputs) in size.

For example, if the input to a regular ConvNet is a 32x32x3 image , a simple 2x2 convolutional layer with regular stride and 12 feature maps would produce a 30x30x12 output. By stacking multiple convolutional layers (and possibly, pooling layers) on top of each other, it becomes possible to scale down the original input to - for example - 10x10x64.

This allows convolutional neural networks to learn a hierarchy of features, from the more detailed ones (at the start of the network) to the clearer ones (towards the tail of the network). 

**Expansive Path**

The contracting network is used, but should be extended with an upsampling network to reconstruct an output at a specific resolution. This network is called the expansive path.

**The Need for Skip Connections**

 In U-Net, such a thing is achived through skip connections, which carry the output of each level in the contracting network to the corresponding level in the upsampling network. It provides upsampling is much more better by combining the high-level information about the input image (which comes from the end of the contracting network) with the high-level information about the level.

 ## Contracting Network Details:

* The contracting part is composed of multiple building blocks of convolutional layers, the so-called convolutional blocks.

* The number of convolutional blocks is 5.

* The output of each convolutional block, except for the last, is used as a skip connection, so in the five-block setting, we have four skip connections in total.

* Each convolutional block in the U-Net contracting path is composed of two convolutional layers with a 3x3 kernel size. They are not padded and each convolutional layer followed by a rectified linear unit (ReLU). Recall that when a 3x3 kernel is used with stride 1, image height and width reduce by 2 pixels each.

* Between each convolutional block, a max pooling operation is performed. It uses a 2x2 pool size and a stride 2 for downsampling. This means that the input image is cut in half width and height wise. This is compensated with a doubling in feature map size.

* Through the contracting path, an input image is downsampled, and a lot of features are learned. These can now be used for upsampling, both using the output of the contractive path and the skip connections.

## Expansive Network Details:

*  The expansive path contains an equal number of blocks with the contracting path, but these are upsampling blocks. The general procedure for these upsampling blocks is as follows:

  - Upsampling is performed on the feature maps at a specific level, to effectively go one level upwards in terms of feature map size. Upsampling also decreases the number of feature maps by 2 (Contracting increases by 2). 
  -The skip connection has the same number of feature maps as the upsampled input.
  - Both are concatenated, with the skip connection first followed by the upsampled input. The number of feature maps is doubled.
  - With two convolutional layers, the feature maps are downsampled again and the number of feature maps is reduced to the original number. So, for example, when the concatenation yields 512x2 = 1024 feature maps, the Conv layers yield 512 again.

* For upsampling, upsampling layers with a 2x2 kernel and stride 2 are used. The convolutional layers are once again with 3x3 kernel size and stride 1, followed by ReLU activation.

* It is end up with an upsampled image with an initial amount of feature maps. 

At the end the model architecture:

* The last layer is a convolution layer with 1 filter (since we have only one class) of size 1x1 kernel to reduce the number of feature maps. The width and height remain the same, but the feature maps are then reduced.

* Activation Function is sigmoid, since we examine an binary classification (roof or not) and loss function is binary cross entropy. The chosen optimizer is Adam.

Note : For both contracting and expansive parts batch normalization and dropout regularization teqniques are used.

**ALL OPERATION LAYER BY LAYER CAN BE SEEN IN THE OUTPUT OF MODEL.SUMMARAY()**




















```python
from keras.models import *
from keras.layers import *
from keras.optimizers import *
from keras.callbacks import ModelCheckpoint, LearningRateScheduler
from keras import backend as keras
#from keras.preprocessing.image import ImageDataGenerator
import skimage.io as io
```

```python
def conv2d_block(input_img, n_filters, kernel_size=3, batchnorm=True):
    # first layer
    x = Conv2D(filters=n_filters, kernel_size=(kernel_size, kernel_size), kernel_initializer="he_normal",
               padding="same")(input_img)
    if batchnorm:
        x = BatchNormalization()(x)
    x = Activation("relu")(x)
    # second layer
    x = Conv2D(filters=n_filters, kernel_size=(kernel_size, kernel_size), kernel_initializer="he_normal",
               padding="same")(x)
    if batchnorm:
        x = BatchNormalization()(x)
    x = Activation("relu")(x)
    return x
```

```python
def get_unet(input_img, n_filters=16, dropout=0.5, batchnorm=True):
    # contracting path
    c1 = conv2d_block(input_img, n_filters=n_filters*1, kernel_size=3, batchnorm=batchnorm)
    p1 = MaxPooling2D((2, 2)) (c1)
    p1 = Dropout(dropout*0.5)(p1)

    c2 = conv2d_block(p1, n_filters=n_filters*2, kernel_size=3, batchnorm=batchnorm)
    p2 = MaxPooling2D((2, 2)) (c2)
    p2 = Dropout(dropout)(p2)

    c3 = conv2d_block(p2, n_filters=n_filters*4, kernel_size=3, batchnorm=batchnorm)
    p3 = MaxPooling2D((2, 2)) (c3)
    p3 = Dropout(dropout)(p3)

    c4 = conv2d_block(p3, n_filters=n_filters*8, kernel_size=3, batchnorm=batchnorm)
    p4 = MaxPooling2D(pool_size=(2, 2)) (c4)
    p4 = Dropout(dropout)(p4)
    
    c5 = conv2d_block(p4, n_filters=n_filters*16, kernel_size=3, batchnorm=batchnorm)
    
    # expansive path
    u6 = Conv2DTranspose(n_filters*8, (3, 3), strides=(2, 2), padding='same') (c5)
    u6 = concatenate([u6, c4])
    u6 = Dropout(dropout)(u6)
    c6 = conv2d_block(u6, n_filters=n_filters*8, kernel_size=3, batchnorm=batchnorm)

    u7 = Conv2DTranspose(n_filters*4, (3, 3), strides=(2, 2), padding='same') (c6)
    u7 = concatenate([u7, c3])
    u7 = Dropout(dropout)(u7)
    c7 = conv2d_block(u7, n_filters=n_filters*4, kernel_size=3, batchnorm=batchnorm)

    u8 = Conv2DTranspose(n_filters*2, (3, 3), strides=(2, 2), padding='same') (c7)
    u8 = concatenate([u8, c2])
    u8 = Dropout(dropout)(u8)
    c8 = conv2d_block(u8, n_filters=n_filters*2, kernel_size=3, batchnorm=batchnorm)

    u9 = Conv2DTranspose(n_filters*1, (3, 3), strides=(2, 2), padding='same') (c8)
    u9 = concatenate([u9, c1], axis=3)
    u9 = Dropout(dropout)(u9)
    c9 = conv2d_block(u9, n_filters=n_filters*1, kernel_size=3, batchnorm=batchnorm)
    
    outputs = Conv2D(1, (1, 1), activation='sigmoid') (c9)
    model = Model(inputs=input_img, outputs=outputs)
    return model
```

```python
input_img = Input((256,256, 4), name='img')
model = get_unet(input_img, n_filters=16, dropout=0.05, batchnorm=True)
model.compile(optimizer=Adam(), loss="binary_crossentropy", metrics=["accuracy"])
model.summary()
```

```python
epochs=10
steps_per_epoch=300 # normally -->(Total Number of Training Samples) / (Batch Size), but for data augmentation it can be changed, since for every 
                    # steps_for_epoch it will create an augmented images of the samples iteratively. 
                    
history = model.fit(myGene,steps_per_epoch=steps_per_epoch,epochs=epochs)
```

```python
def saveResult(save_path,results,filenames):
    for i,item in enumerate(results):
        img = item[:,:,0]
        img_name = filenames[i]
        io.imsave(os.path.join(save_path, img_name),img)
        
saveResult("./sample_data/test/labels/",(test_2*255).astype(np.uint8), file_names)
```
